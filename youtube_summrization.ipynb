```python
from youtube_transcript_api import YouTubeTranscriptApi
import google.generativeai as genai
import textwrap
from IPython.display import display
from IPython.display import Markdown


video = input("Please Enter the youtube URL")
data = { 'url': video }
try:
    resp = YouTubeTranscriptApi.get_transcript(data['url'].split("=")[-1])
except:
   print("Subtitle is off for this video")

youtube_transcript = ' '.join([i.get('text','') for i in resp])
print("Extracted Youtube Text")

genai.configure(api_key='AIzaSyDGfyDeuQ7EN0jtzqEeIX8IZBBSoitpIfU')
model = genai.GenerativeModel('gemini-pro')

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

print("Summarizing.......")

response = model.generate_content(f"Kindly assist me in condensing the content of this video into key points. Include all significant information. Original Text: {youtube_transcript}")
to_markdown(response.text)
```
